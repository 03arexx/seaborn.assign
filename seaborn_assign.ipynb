Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python visualization library that works well with pandas DataFrames and integrates closely with matplotlib.

1.Scatter Plot (sns.scatterplot):
Use: To visualize the relationship or correlation between two continuous variables. Each point represents an observation in the dataset. It helps in identifying patterns, trends, and outliers.

2.Line Plot (sns.lineplot):
Use: To display data points connected by lines, which is useful for visualizing time series data or trends over continuous variables. It helps in understanding the direction and strength of relationships between variables.

3.Bar Plot (sns.barplot):
Use: To show the distribution of a categorical variable or to compare the mean values of different categories. It's helpful for understanding the central tendency and comparing groups.

4.Box Plot (sns.boxplot):
Use: To depict the distribution of a continuous variable across different categories. It highlights the median, quartiles, and potential outliers, providing insights into the spread and skewness of the data.

5.Heatmap (sns.heatmap):
Use: To represent data in a matrix format, where the individual values are represented by colors. It is often used to show the correlation matrix, frequency counts, or any data with two dimensions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
flights = sns.load_dataset('flights')

# Pivot the dataset to get a matrix of values for the heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)
plt.title('Number of Passengers per Month')
plt.show()
```
